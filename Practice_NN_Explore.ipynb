```
# Importing Libraries 
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

**Coding Stepback Practice**

```
my_layer = Dense(units=1, input_shape=[1]) 
model = Sequential([my_layer])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
```

```
print(model.predict([10.0]))
```

**Printing weights of a layer**

```
print(my_layer.get_weights())
```

**Experimenting with 2 layer, multi-neuron network**

```
# Building the Network
my_layer1 = Dense(units=2, input_shape=[1])
my_layer2 = Dense(units=1)

# Creating the experimental model
model_exper = Sequential([my_layer1, my_layer2])
model_exper.compile(optimizer='sgd', loss='mean_squared_error')
```

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model_exper.fit(xs, ys, epochs=500)
```

```
print(my_layer1.get_weights())
```

```
print(my_layer2.get_weights())
```

```
value_to_predict = 10.0

# Defining the layer manually, by assigning the learned weights and then making a prediction out of it.
layer1_w1 = (my_layer1.get_weights()[0][0][0])
layer1_w2 = (my_layer1.get_weights()[0][0][1])
layer1_b1 = (my_layer1.get_weights()[1][0])
layer1_b2 = (my_layer1.get_weights()[1][1])

layer2_w1 = (my_layer2.get_weights()[0][0])
layer2_w2 = (my_layer2.get_weights()[0][1])
layer2_b = (my_layer2.get_weights()[1][0])

neuron1_layer1_output = (layer1_w1*value_to_predict)+layer1_b1
neuron2_layer1_output = (layer1_w2*value_to_predict)+layer1_b2

neuron3_layer2_output = ((neuron1_layer1_output*layer2_w1)+(neuron2_layer1_output*layer2_w2))+layer2_b

print(neuron3_layer2_output)
```
